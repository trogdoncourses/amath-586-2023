```julia
using Plots, LinearAlgebra, LaTeXStrings, Printf
```

```julia
h = 0.0001
x = h:h:1-h
m = convert(Int64,1/h)-1;
A = SymTridiagonal(fill(-2.0,m),fill(1.0,m-1))/h^2;
E = eigen(A);
```

```julia
λtrue = [ -j^2*pi^2 for j = m:-1:1];
```

```julia
sm = convert(Int64,floor(sqrt(m)/4))
plot(reverse(E.values - λtrue)[1:sm], legend = false, yaxis = [0,2e-2], markershape = :diamond )
```

```julia
s = j -> x -> sin(j*pi*x)*sqrt(2)
```

```julia
j = 1
q = E.vectors[:,end-j+1]/sqrt(h)
q *= sign(q[1])
p = plot(x,q,label = latexstring(@sprintf("q_{%i}",j)), aspect_ratio = .1, yaxis = [-2,2], framestyle = :box, legend = :outertop)
plot!(x,map(s(j),x),label = latexstring(@sprintf("\\,\\sqrt{2}\\sin({%i} \\pi x)",j)))
```
